## Задание 3: первичный анализ

Выведем статистику для каждого файла по образцу:

```python
def analyze_file(filename):

  # 1. Разбиваем содержание файла на строки
  with open(filename, 'r', encoding='utf-8') as file:
    full_lines = [line.rstrip('\n') for line in file]

  # 2. Считаем статистику
  total_lines = 0         # общее количество строк
  total_words = 0         # общее количество слов
  total_chars = 0         # общее количество символов
  non_empty_lines = 0     # количество непустых строк
  non_empty_lines_list = list()

  # Ваш код здесь: реализуйте логику для получения таких статистик
  for line in full_lines:
    total_lines += 1
    total_chars += len(line)
    words_in_line = line.split()
    total_words += len(words_in_line)

    if line.strip():
      non_empty_lines += 1
      non_empty_lines_list.append(line)

  # 3. Анализ длин строк
  if full_lines:
    max_line_length = max([len(line) for line in full_lines])
    non_empty_length = [len(line) for line in full_lines if line.strip()]
    if non_empty_length:
      min_line_length = min(non_empty_length)

  # Выводим результат на экран для каждого файла в таком формате:
  print(f"\n Результаты анализа {filename}\n")
  print(f"Всего строк: {total_lines}")
  print(f"Непустых строк: {non_empty_lines}")  # ИСПРАВЛЕНО: выводим число, а не список
  print(f"Всего слов: {total_words}")
  print(f"Всего символов: {total_chars}")
  print(f"Макс. длина строки: {max_line_length}")
  print(f"Мин. длина строки (среди непустых): {min_line_length}")

analyze_file("twitter.txt")
analyze_file("twitterconv.txt")
analyze_file("movie.txt")
analyze_file("bncsplit.txt")
analyze_file("bnc.txt")
```

## Задание 4: чистка текста

```python
import re

def cleaned_text(filename):
    cleaned_lines = [] # чистим текст построчно (сохраняем каждую очищенную строку в список)
    all_cleaned_words = [] # также сохраняем все слова после очистки в список

    # Ваш код здесь: реализуйте очистку с помощью цикла for и RegEx
    # 1. Используйте strip для удаления лишних пробелов из строки
    # 2. Очищаем строку: оставляем только буквы, цифры и пробелы
    # 3. Убираем лишние пробелы
    # 4. Приводим к нижнему регистру
    # 5. Собираем все слова после очистки в единый список
    #    - не фиксируем слова короче 1 символа

    with open(filename, "r", encoding="utf-8") as file:
        full_lines = [line.strip() for line in file]

    for line in full_lines:
        line = re.sub(r"[^\w\s]|_", " ", line)
        line = re.sub(r"\s+", " ", line)
        line = line.lower()
        cleaned_lines.append(line)

        words = line.split()
        new_words = [word for word in words if len(word) >= 1]
        all_cleaned_words.extend(new_words)

    # Выводим результат для каждого текста по образцу
    print(f"Пример очищенной строки: {cleaned_lines[0][:100]}...")
    print(f"Всего очищенных слов: {len(all_cleaned_words)}")

    return filename, all_cleaned_words

filename1, all_cleaned_words1 = cleaned_text("twitter.txt")
filename2, all_cleaned_words2 = cleaned_text("twitterconv.txt")
filename3, all_cleaned_words3 = cleaned_text("movie.txt")
filename4, all_cleaned_words4 = cleaned_text("bncsplit.txt")
filename5, all_cleaned_words5 = cleaned_text("bnc.txt")
```

## Задание 5: статистика

Часть 1: получаем статистику распределений длин слов

```python
import re

def cleaned_text(filename):
    cleaned_lines = []
    all_cleaned_words = []

    with open(filename, "r", encoding="utf-8") as file:
        full_lines = [line.strip() for line in file]

    for line in full_lines:
        line = re.sub(r"[^\w\s]|_", " ", line)
        line = re.sub(r"\s+", " ", line)
        line = line.lower()
        cleaned_lines.append(line)

        words = line.split()
        new_words = [word for word in words if len(word) >= 1]
        all_cleaned_words.extend(new_words)

    if cleaned_lines:
        print(f"Пример очищенной строки: {cleaned_lines[0][:100]}...")
    else:
        print("Пример очищенной строки: (файл пустой)")

    print(f"Всего очищенных слов: {len(all_cleaned_words)}")

    return filename, all_cleaned_words

def statistics_final(filename, all_cleaned_words):
    total_words = len(all_cleaned_words)
    word_lengths = [len(word) for word in all_cleaned_words]
    total_chars = sum(word_lengths)

    word_length_count = {}
    for length in word_lengths:
        if length in word_length_count:
            word_length_count[length] += 1
        else:
            word_length_count[length] = 1

    print(f"Отчет по {filename}:")
    print(f"Слов после очистки: {total_words}")
    print(f"Символов после очистки: {total_chars}")
    print("Распределение длин слов:")
    for length in sorted(word_length_count.keys()):
        print(f"  {length} букв: {word_length_count[length]} слов")
    print()

# Основной код
filename1, all_cleaned_words1 = cleaned_text("twitter.txt")
filename2, all_cleaned_words2 = cleaned_text("twitterconv.txt")
filename3, all_cleaned_words3 = cleaned_text("movie.txt")
filename4, all_cleaned_words4 = cleaned_text("bncsplit.txt")
filename5, all_cleaned_words5 = cleaned_text("bnc.txt")

statistics_final(filename1, all_cleaned_words1)
statistics_final(filename2, all_cleaned_words2)
statistics_final(filename3, all_cleaned_words3)
statistics_final(filename4, all_cleaned_words4)
statistics_final(filename5, all_cleaned_words5)
```

Часть 2: частотность слов

```python
def frequency(filename, all_cleaned_words):
  # Считаем частотность слов (по аналогии с подсчетом word_length_count)
  word_frequency = {}
  # Ваш код здесь:
  # проходим по all_cleaned_words
  for word in all_cleaned_words:
    # если слово зафиксировано в word_frequency, то
    if word in word_frequency:
        word_frequency[word] += 1
    # иначе
    else:
        word_frequency[word] = 1

# Находим топ-10 самых частых слов
# Начинаем с формирования кортежей вида (частотность слова, лексема)
word_count_pairs = []
for word, count in word_frequency.items():
    word_count_pairs.append((count, word))

# Сортируем вручную через вложенные циклы (пузырьковая сортировка)
# Стефания: пузырьковая сортировка шла минут 30 и еще не заканчивалась, поменяла на:

word_count_pairs.sort(reverse=True)

print("Топ-10 самых частых слов:")
for i in range(min(10, len(word_count_pairs))):
    count, word = word_count_pairs[i]
    print(f"  {i+1}. '{word}': {count} раз")

# Анализ уникальных слов
# Ваш код здесь: создайте множество unique_words
unique_words = set(all_cleaned_words)

print(f"\nВсего уникальных слов: {len(unique_words)}")

frequency(filename1, all_cleaned_words1)
frequency(filename2, all_cleaned_words2)
frequency(filename3, all_cleaned_words3)
frequency(filename4, all_cleaned_words4)
frequency(filename5, all_cleaned_words5)
```
